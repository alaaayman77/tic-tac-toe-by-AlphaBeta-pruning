```python
import numpy as np
from random import choice
from math import inf


board = np.zeros((3,3))
# print(board)

#display game board
def GameBoard(board):

    #game rule
    #row by row
    for x in range(0,3):
        #element by element
        for y in range(0,3):
            #access values in chars dic by using key y
            # 0 -> _
            # 1 -> X (player)
            # -1 -> O (AI)
            if board [x][y] == 0:
                print("_" ,end=' ')
            elif board[x][y] == 1:
                print('X' ,end=' ')
            else:
                print('O',end=' ')
        print()

    print("============================")

#clear board 
# def ClearBoard(board):
#         board[:] = 0

#check win
def WinningPlayer(board,player):
    conditions=[
        #horizontal 
        [board[0][0],board[0][1],board[0][2]],
        [board[1][0],board[1][1],board[1][2]],
        [board[2][0],board[2][1],board[2][2]],
        #vertical
        [board[0][0],board[1][0],board[2][0]],
        [board[0][1],board[1][1],board[2][1]],
        [board[0][2],board[1][2],board[2][2]],
        #cross
        [board[0][0],board[1][1],board[2][2]],
        [board[0][2],board[1][1],board[2][0]]
    ]

#if there is [1,1,1] or [-1,-1,-1] in the same inner array like [b[0][0],b[0][1],b[0][2]]
    if [player,player,player] in conditions:
        return True
    return False


#print game result
def printResult(board):
    if WinningPlayer(board,1):
        print("Congratulations! You won the game.")
    elif WinningPlayer(board,-1):
        print("Sorry. The AI has won this time.")
    else:
        print("It's a draw.")


#check blank positions
def blanks(board):
    #array for to store in it blank position
    blank =[]
    # x -> index of each row 0 - 1 - 2 - 3
    # row -> element of this row [0,0,0]
    for x,row in enumerate(board):
        # y -> index of each element in row 
        # col -> value of element in col
        for y,col in enumerate(row):
            #x,y -> (0 , 0) - (0 , 1) - (0 , 2)
            if board[x][y] == 0:
                blank.append([x,y])
    return blank
#Is FULL
# def boardFull(board):
#     if len(blanks(board)) == 0:
#         return True
#     return False

#Set player move on the board for AI(-1) or human(1)
def setMove(board,x,y,player):
    board[x][y] = player

def playerMove(board):
    flag = True
    moves = {
        1:[0,0],
        2:[0,1],
        3:[0,2],
        4:[1,0],
        5:[1,1],
        6:[1,2],
        7:[2,0],
        8:[2,1],
        9:[2,2]
        }
    while flag:
        try:
            move = int(input("Enter number between 1-9: "))
            if move < 1 or move > 9:
                print("Invalid Move!")
            elif  not (moves[move] in blanks(board)):
                print("Invalid Move")
            else:
                #moves is dict each key has array so we indecent first key and second array element
                #move -> key 0->(X)first element in array 1 ->(Y) second element
                setMove(board ,moves[move][0],moves[move][1],1)
                GameBoard(board)
                flag=False
                #when using input is not int
        except(KeyError,ValueError):
            print("Enter number")
    

#game state
def evaluate(board):
    if WinningPlayer(board,1):
        return 1
    elif WinningPlayer(board,-1):
        return -1
    else:
        return 0 

#------------------------------
# def heuristic(board):
#     return evaluate(board)
#---------------------------


# def alphaBetaPruning(board,depth,alpha,beta,player):

#     if depth == 9 or WinningPlayer(board,-1) or WinningPlayer(board,1):
#         return 0
    
#     if player == 1: #(max - human)
#         best = MAX
#         for i in range(0,9):
#             v = alphaBetaPruning(board,depth+1,alpha,beta,-player)
#             best = max(best,v)
#             alpha = max(alpha,best)

#             if beta <= alpha:
#                 break
#         return best
    
#     else:
#         best = MIN
#         for i in range(0,9):
#             v = alphaBetaPruning(board,depth+1,alpha,beta,player)
#             best = min(best,v)
#             beta = min(beta,best)

#             if beta <= alpha:
#                 break
#         return best

def abminimax(board, depth, alpha, beta, player):
    row = -1
    col = -1
    if depth == 0 or WinningPlayer(board,-1) or WinningPlayer(board,1):
        return [row, col, evaluate(board)]
    
    else:
        for cell in blanks(board):
            setMove(board, cell[0], cell[1], player)
            score = abminimax(board, depth - 1, alpha, beta, -player)
            if player == 1:
                # X is always the max player

                #score[row ,col ,player]
                if score[2] > alpha:
                    alpha = score[2]
                    row = cell[0]
                    col = cell[1]


            else:
                if score[2] < beta:
                    beta = score[2]
                    row = cell[0]
                    col = cell[1]

            #trice all empty cells to find optimal solution
            setMove(board, cell[0], cell[1], 0)
            
            if alpha >= beta:
                break

        if player == 1:
            return [row, col, alpha]
        
        else:
            return [row, col, beta]

# AI
def o_comp(board):
    if len(blanks(board)) == 9:
        x = choice([0, 1, 2])
        y = choice([0, 1, 2])
        setMove(board, x, y, -1)
        GameBoard(board)

    else:
        result = abminimax(board, len(blanks(board)), -inf, inf, -1)
        setMove(board, result[0], result[1], -1)
        GameBoard(board)

# def x_comp(board):
#     if len(blanks(board)) == 9:
#         x = choice([0, 1, 2])
#         y = choice([0, 1, 2])
#         setMove(board, x, y, 1)
#         GameBoard(board)

#     else:
#         result = abminimax(board, len(blanks(board)), -inf, inf, 1)
#         setMove(board, result[0], result[1], 1)
#         GameBoard(board)


def makeMove(board, player):
        if player == 1:
            playerMove(board)

        else:
            o_comp(board)

def ClearBoard(board):
    board[:]=0

def pvc():
    while True:
        try:
            order = int(input('Enter to play 1st or 2nd: '))
            if not (order == 1 or order == 2):
                print('Please pick 1 or 2')
            else:
                break
        except(KeyError, ValueError):
            print('Enter a number')

    ClearBoard(board)
    if order == 2:
        #AI
        currentPlayer = -1
    else:
        #Human
        currentPlayer = 1
                    # T->board full
                    #F->empty       
    while not (len(blanks(board))==0 or  WinningPlayer(board,-1) or  WinningPlayer(board,1)):
        makeMove(board, currentPlayer)
        currentPlayer *= -1

    printResult(board)


pvc()
```
